Bubble sort algorithm and implementation

```python
arr=[5,3,8,4] #i create a list with values using arr variable
n=len(arr) # length of array n=4
for i in range(n): #outter loop run 4 times 0,1,2,3
  for j in range(n-1): #inner loop this loop compares adjacent elements j goes index 0,1,2
    if arr[j]>arr[j+1]: # It check if the left element is greater than right element means 5>3 swap
      arr[j],arr[j+1]=arr[j+1],arr[j] # swap two elements
print('bubble sort array:',arr) # it display the output
```

Selection sort algorthim and implementation

```python
arr=[7,12,9,11,3] # I create a list with values using arr variable
for i in range(len(arr)): # outter loop loops runs 0,1,2,3,4 len=5
   for j in range(i+1,len(arr)): # inner loop, it compare the current element with all elements
    if arr[i]>arr[j]: # it find smaller element, if yes means it they are wrong
      arr[i],arr[j]=arr[j],arr[i] # if smallest element is found it immediately swap them
print("Selection sort array:",arr) # it display the output
```

Quick Sort algorithm and implementation

```python
arr=[10,3,55,9,11] # I create a list with values using arr variable
pivot=arr[0] # it choose the element as the pivot for compare all values
left=[] # I create a empty list with left variable to store smaller values
right=[] # i create a empty list with right variable to store large values
for i in range(1,len(arr)): # loop through all value expect the pivot element
  if arr[i]<pivot: # if element is smaller than pivot, moves to the left list otherwise it moves to right
    left.append(arr[i])
  else:
    right.append(arr[i])
arr=left+[pivot]+right #it combine smaller values, pivot and large values
arr.sort() # most of cases it is in unsorted array, I use a sort method
print("Quick sort:",arr) # finally it display the
```
